Read australian_users_items.json that contains values in pyhton dictionary format. Filter out the values by fetching user_id, item_id, item_name and playtime_forever and save it in csv file output_australian_users_items.csv

```python
import csv
import ast
import json
import pandas as pd

csv_columns = ['user_id','item_id','item_name','playtime_forever']

with open('C:\\Users\\Kriti\\Desktop\\CMPE256\\CMPE256_Project\\australian_users_items.json', encoding='utf-8') as data_file:
    
    with open('C:\\Users\\Kriti\\Desktop\\CMPE256\\CMPE256_Project\\output_australian_users_items.csv', 'w', encoding='utf-8', newline='') as csv_file:
        line = data_file.readline()
        writer = csv.DictWriter(csv_file, fieldnames=csv_columns)
        writer.writeheader()
        while line:
            json_data = (ast.literal_eval(line))
            line = data_file.readline()
            for item in json_data["items"]:
                new_dict_data={'user_id':json_data["user_id"],'item_id':item['item_id'],'item_name':item['item_name'],'playtime_forever':item['playtime_forever']}
            
                writer.writerow(new_dict_data)
                line = data_file.readline()
```

Read australian_user_reviews.json that contains values in pyhton dictionary format. Filter out the values by fetching user_id, item_id, recommend and save it in csv file output_australian_user_reviews.csv

```python
import csv
import ast
import json
import pandas as pd

csv_columns = ['user_id','item_id','recommend']

with open('C:\\Users\\Kriti\\Desktop\\CMPE256\\CMPE256_Project\\australian_user_reviews.json', encoding='utf-8') as data_file:
    
    with open('C:\\Users\\Kriti\\Desktop\\CMPE256\\CMPE256_Project\\output_australian_user_reviews.csv', 'w', encoding='utf-8', newline='') as csv_file:
        line = data_file.readline()
        writer = csv.DictWriter(csv_file, fieldnames=csv_columns)
        writer.writeheader()
        while line:
            json_data = (ast.literal_eval(line))
            line = data_file.readline()
            for review in json_data["reviews"]:
                new_dict_data={'user_id':json_data["user_id"],'item_id':review['item_id'],'recommend':review['recommend']}
            
                writer.writerow(new_dict_data)
                line = data_file.readline()

```
